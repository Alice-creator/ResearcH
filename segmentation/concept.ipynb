```python
import numpy as np
print("=== T·∫†I SAO D√ôNG EXPONENTIAL? ===")
print("w(x) = wc(x) + w0 * exp(-(d1(x) + d2(x))¬≤/(2œÉ¬≤))")
print()

print("üéØ L√ù DO CH·ªåN EXPONENTIAL:")
print("1. SMOOTH DECAY: Gi·∫£m m∆∞·ª£t m√† theo kho·∫£ng c√°ch")
print("2. LOCALIZED EFFECT: Ch·ªâ ·∫£nh h∆∞·ªüng v√πng g·∫ßn boundary")
print("3. STRONG EMPHASIS: TƒÉng weight m·∫°nh cho pixels quan tr·ªçng")
print()

print("üìä SO S√ÅNH C√ÅC H√ÄNG S·ªê:")
import numpy as np
distances = [1, 2, 3, 5, 10]
w0, sigma = 10, 5

for d in distances:
    exp_weight = w0 * np.exp(-(d**2)/(2*sigma**2))
    linear_weight = max(0, w0 - d)  # Linear decay
    print(f"Distance {d}: exp={exp_weight:.2f} vs linear={linear_weight:.2f}")

print()
print("üî• EXPONENTIAL ADVANTAGES:")
print("‚Ä¢ Decay nhanh ‚Üí ch·ªâ ·∫£nh h∆∞·ªüng v√πng g·∫ßn boundary")
print("‚Ä¢ Smooth transition ‚Üí kh√¥ng c√≥ sharp edges")
print("‚Ä¢ Gaussian-like shape ‚Üí natural weighting")
print("‚Ä¢ Hyperparameter œÉ control ‚Üí flexible tuning")
```
